```python
import numpy as np
import matplotlib.pyplot as plt
import random
from PIL import Image, ImageOps
import math
```

# Loading a png file as Configuration Space

```python
cspace = Image.open('cspace.png')
cspace = ImageOps.grayscale(cspace)
np_cspace = np.array(cspace)
np_cspace = ~np_cspace
np_cspace[np_cspace > 0] = 1
```

```python
class TreeNode():
    def __init__(self, x, y, parent=None):
        self.x = x
        self.y = y
        self.parent = parent
        self.children = []
        self.in_path = False
        
```

```python
# I shoud standardise the format whether to use oop or tuples or list lol
class RRT():
    def __init__(self, start ,goal, cspace, step_size):
        self.start = start
        self.goal = goal
        self.cspace = cspace
        self.step_size = step_size
        self.tree = TreeNode(*start)
        self.nodes = [self.tree]

    def show(self):
        plt.imshow(np_cspace, cmap="binary", origin='lower')

        # Get path nodes
        path_nodes = [node for node in self.nodes if node.in_path == True]
        other_nodes = [node for node in self.nodes if node.in_path == False]
        
        xs_path = [node.x for node in path_nodes]
        ys_path = [node.y for node in path_nodes]

        xs_other = [node.x for node in other_nodes]
        ys_other = [node.y for node in other_nodes]

        # plot path red and other nodes blue
        plt.scatter(xs_path, ys_path, c='red', s=10)
        plt.scatter(xs_other, ys_other, c='yellow', s=10)
        
        plt.scatter(self.start[0], self.start[1], c='green', s=50, label='Start')
        plt.scatter(self.goal[0], self.goal[1], c='blue', s=50, label='Goal')
        plt.show()

    def distance(self, node_1, node_2):
        return math.sqrt((node_1[0]-node_2[0])**2 + (node_1[1]-node_2[1])**2)

    # This is super inefficient and can be improved
    def find_nearest_node(self, cur_node):
        nearest_node = None
        nearest_distance = math.inf
        for potential_node in self.nodes:
            cur_distance = self.distance(cur_node, (potential_node.x, potential_node.y))
            if  cur_distance < nearest_distance:
                nearest_node = potential_node
                nearest_distance = cur_distance
        return nearest_node, nearest_distance
            
        

    def add_node(self):
        # Generate a sample point
        sample_point = (np.random.uniform(0,np_cspace.shape[1]), np.random.uniform(0, np_cspace.shape[0]))
        # Get the vector from nearest point to sample point 
        nearest_node, nearest_distance = self.find_nearest_node(sample_point)
        # Get unit vector from nearest node to sample_point
        unit_vector = [(sample_point[0]-nearest_node.x)/nearest_distance, (sample_point[1]-nearest_node.y)/nearest_distance]
        # Generate the new node
        d = random.randrange(0, self.step_size)
        new_node = (nearest_node.x + d*unit_vector[0], nearest_node.y + d*unit_vector[1])
        new_node = (math.floor(new_node[0]), math.floor(new_node[1]))
        # Check for out of bounds
        if new_node[0] < 0 or new_node[0] >= self.cspace.shape[1]:
            return
        if new_node[1] < 0 or new_node[1] >= self.cspace.shape[0]:
            return
        # Check for collision
        if self.cspace[new_node[1], new_node[0]] == 1:
            return
        # Add new node to list of nodes if no collision
        new_node = TreeNode(*new_node, nearest_node)
        self.nodes.append(new_node)
        return new_node

    def search(self, n_iter):
        cur_iter = 0
        while cur_iter < n_iter:
            # Generate a new node
            new_node = self.add_node()
            # Check if its collision
            if new_node is None:
                continue
            # Check if goal is reached
            if self.distance((new_node.x, new_node.y), self.goal) <= 80:
                while new_node is not None:
                    new_node.in_path = True
                    new_node = new_node.parent
                return
            cur_iter += 1
            
        
        
        
        
```

```python
rrt = RRT([50,50],[1400,900],np_cspace,80)
```
